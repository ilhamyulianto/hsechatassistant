```python
from PyPDF2 import PdfReader
import re
```

```python
pdf_path = "materials\service-gdc-gdcovop-2012330878-2012330878.pdf"
```

```python
#read

reader = PdfReader(pdf_path)
raw_text = ""
for page in reader.pages:
    text = page.extract_text()
    if text:
        raw_text += text + "\n"
```

```python
#clean and remove 

clean_text = raw_text
clean_text = re.sub(r'\n+', '\n', clean_text)
clean_text = re.sub(r'\s{2,}', ' ', clean_text)
clean_text = re.sub(r'Page\s*\d+', '', clean_text, flags=re.IGNORECASE)
```

```python
with open("materials\cleaned.txt", "w", encoding="utf-8") as f:
    f.write(clean_text)

print(clean_text[:800])
```
