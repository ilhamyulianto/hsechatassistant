```python
from sentence_transformers import SentenceTransformer
import numpy as np
import re
```

```python
model = SentenceTransformer("intfloat/multilingual-e5-base")
```

```python
file_path = "materials\chunks.txt"
with open(file_path, "r", encoding="utf-8") as f:
    text = f.read()
    
#split
chunks = re.split(r"\[CHUNK \d+\]\n", text)
chunks = [c.strip() for c in chunks if c.strip()]
```

```python
#embedding
embeddings = model.encode(chunks, show_progress_bar=True, convert_to_numpy=True)
```

```python
np.save("chunk_embeddings.npy", embeddings)
with open("materials\postembedding.txt", "w", encoding="utf-8") as f:
    for c in chunks:
        f.write(c + "\n---\n")
        
print(f"result {len(embeddings)} embeddings.")
print(f"{embeddings.shape[1]} dimensions.")
```
